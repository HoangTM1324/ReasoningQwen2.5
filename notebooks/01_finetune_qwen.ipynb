```python
!pip install -r requirements.txt -q

# Import từ src của chúng ta
from src import (
    load_qlora_model,
    get_processed_dataset,
    setup_peft_model,
    get_training_args
)
from transformers import Trainer

# Config đường dẫn
MODEL_NAME = "Qwen/Qwen2.5-0.5B-Instruct"
DATASET_NAME = "open-thoughts/OpenThoughts-114k"
OUTPUT_DIR = "./qwen-finetuned-output"
```

```python
print("Loading model & tokenizer...")
model, tokenizer = load_qlora_model(MODEL_NAME)
```

```python
print("Processing dataset...")
SYS_PROMPT = "You are a helpful assistant capable of complex reasoning."

dataset = get_processed_dataset(DATASET_NAME, tokenizer, SYS_PROMPT)
print(f"Dataset size: {len(dataset)}")
```

```python
print("Setting up Training...")
model = setup_peft_model(model)
training_args = get_training_args(OUTPUT_DIR)

trainer = Trainer(
    model=model,
    train_dataset=dataset,
    args=training_args,
)

trainer.train()
```
